```python
# Cell 1
from pathlib import Path
import json
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

file_6432 = "KTH_results/both64_500_64_ctx64x64_fov32x32/plots/history_context64x64_fovea32x32_fold1.json"
file_3216 = "KTH_results/both32_500_64_ctx32x32_fov16x16/plots/history_both32_ctx32x32_fov16x16_fold1.json"
out_acc = "accuracy_plot.png"
out_loss = "loss_plot.png"
```

```python
# Cell 2
def load_hist(fp):
    with open(fp, "r", encoding="utf-8") as f:
        d = json.load(f)
    return {k: np.asarray(v, dtype=float) for k, v in d.items()}

h6432 = load_hist(file_6432)
h3216 = load_hist(file_3216)
e6432 = np.arange(1, len(h6432["accuracy"]) + 1)
e3216 = np.arange(1, len(h3216["accuracy"]) + 1)
```

```python
# Cell 3
plt.figure(figsize=(9,5))
plt.plot(e3216, h3216["accuracy"], label="KTH(32,16) training", color="red", marker="o", markevery=12, linewidth=2)
plt.plot(e3216, h3216["val_accuracy"], label="KTH(32,16) validation", color="green", marker="o", markevery=12, linewidth=2)
plt.plot(e6432, h6432["accuracy"], label="KTH(64,32) training", color="blue", marker="o", markevery=12, linewidth=2)
plt.plot(e6432, h6432["val_accuracy"], label="KTH(64,32) validation", color="cyan", marker="o", markevery=12, linewidth=2)
plt.xlabel("Number of Epochs")
plt.ylabel("Accuracy")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(framealpha=0.9, loc="upper center", bbox_to_anchor=(0.5, 1.12), ncol=2)
axins = inset_axes(plt.gca(), width="36%", height="45%", loc="lower right", borderpad=2)
axins.plot(e3216, h3216["accuracy"], color="red", marker="o", markevery=12, linewidth=2)
axins.plot(e3216, h3216["val_accuracy"], color="green", marker="o", markevery=12, linewidth=2)
axins.plot(e6432, h6432["accuracy"], color="blue", marker="o", markevery=12, linewidth=2)
axins.plot(e6432, h6432["val_accuracy"], color="cyan", marker="o", markevery=12, linewidth=2)
axins.set_xlim(400, 500)
axins.set_ylim(0.9, 1.00)
axins.set_xticks([400, 450, 500])
plt.tight_layout()
plt.savefig(out_acc, dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Cell 4
plt.figure(figsize=(9,5))
plt.plot(e3216, h3216["loss"], label="KTH(32,16) training", color="red", marker="o", markevery=12, linewidth=2)
plt.plot(e3216, h3216["val_loss"], label="KTH(32,16) validation", color="green", marker="o", markevery=12, linewidth=2)
plt.plot(e6432, h6432["loss"], label="KTH(64,32) training", color="blue", marker="o", markevery=12, linewidth=2)
plt.plot(e6432, h6432["val_loss"], label="KTH(64,32) validation", color="cyan", marker="o", markevery=12, linewidth=2)
plt.xlabel("Number of Epochs")
plt.ylabel("Loss")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(framealpha=0.9, loc="upper center", bbox_to_anchor=(0.5, 1.12), ncol=2)
axins = inset_axes(plt.gca(), width="36%", height="45%", loc="upper right", borderpad=2)
axins.plot(e3216, h3216["loss"], color="red", marker="o", markevery=12, linewidth=2)
axins.plot(e3216, h3216["val_loss"], color="green", marker="o", markevery=12, linewidth=2)
axins.plot(e6432, h6432["loss"], color="blue", marker="o", markevery=12, linewidth=2)
axins.plot(e6432, h6432["val_loss"], color="cyan", marker="o", markevery=12, linewidth=2)
axins.set_xlim(400, 500)
axins.set_ylim(0.0, 0.3)
axins.set_yticks([0.1, 0.3])
plt.tight_layout()
plt.savefig(out_loss, dpi=300, bbox_inches="tight")
plt.show()
```
